```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import matplotlib as mpl

mpl.rcParams['font.size'] = 8    # 全局小一点

threads = ['North 1', 'South 2', 'North 3', 'North 4', 'South 5']
phases = {'North 1': {
        'Delay Wait':      (0, 3),    # 21:06:19 → 22
        'Bridge Wait':     (3, 0),    # 获得桥权限立即进门
        'North Gate Wait': (3, 0),
        'Pass North Gate': (3, 2),    # 22 → 24
        'Pass Bridge':     (5, 5),    # 24 → 29
        'South Gate Wait': (10, 0),
        'Pass South Gate': (10, 2),   # 29 → 31
    },

    'South 2': {
        'Delay Wait':      (0, 4),    # 23
        'Bridge Wait':     (4, 0),
        'South Gate Wait': (4, 0),
        'Pass South Gate': (4, 2),    # 23 → 25
        'Pass Bridge':     (6, 5),    # 25 → 30
        'North Gate Wait': (11, 0),
        'Pass North Gate': (11, 2),   # 30 → 32
    },

    'North 3': {
        'Delay Wait':      (0, 5),    # 24
        'Bridge Wait':     (5, 7),    # 24 → 31
        'North Gate Wait': (12, 1),   # 31 → 32
        'Pass North Gate': (13, 2),   # 32 → 34
        'Pass Bridge':     (15, 2),   # 34 → 36
        'South Gate Wait': (17, 0),
        'Pass South Gate': (17, 2),   # 36 → 38
    },

    'North 4': {
        'Delay Wait':      (0, 6),    # 25
        'Bridge Wait':     (6, 13),   # 25 → 38
        'North Gate Wait': (19, 1),   # 38 → 39
        'Pass North Gate': (20, 2),   # 39 → 41
        'Pass Bridge':     (22, 5),   # 41 → 46
        'South Gate Wait': (27, 0),
        'Pass South Gate': (27, 2),   # 46 → 48
    },

    'South 5': {
        'Delay Wait':      (0, 5),    # 24
        'Bridge Wait':     (5, 8),    # 24 → 32
        'South Gate Wait': (13, 0),
        'Pass South Gate': (13, 2),   # 32 → 34
        'Pass Bridge':     (15, 3),   # 34 → 37
        'North Gate Wait': (18, 0),
        'Pass North Gate': (18, 2),   # 37 → 39
    },}      # 你的 phases 字典，省略
color_map = { 
    'Delay Wait'      : '#A9A9B3',  # muted grey
    'Bridge Wait'     : '#F4D35E',  # warm mustard
    'South Gate Wait' : '#F28C28',  # soft orange
    'North Gate Wait' : '#5DA5DA',  # calm blue
    'Pass North Gate' : '#B5D9EA',  # pastel blue
    'Pass South Gate' : '#F7A1A1',  # light salmon
    'Pass Bridge'     : '#73C19A',  # gentle green
    }   # 你的 color_map 字典，省略

fig, ax = plt.subplots(figsize=(11, 4.5))

bar_h = 0.4                     # 条形高度
y_pos = {t: i for i, t in enumerate(threads)}   # 把类别映射到数值 y

for thread in threads:
    y = y_pos[thread]
    for phase, (start, dur) in phases[thread].items():
        if dur == 0:
            continue
        ax.barh(y, dur, left=start, height=bar_h,
                color=color_map[phase], edgecolor='none')

        # -------- 智能放置标签 --------
        txt = phase.replace(' ', '\n')   # 把空格换成换行，可选
        text_len = len(txt.replace('\n', '')) * 0.11  # 估算文字宽度（粗略）
        inside = dur > text_len + 0.6                 # 条形足够宽就放里面

        if inside:
            ax.text(start + dur/2, y,
                    txt, ha='center', va='center',
                    color='white', fontsize=7, linespacing=0.9)
        else:
            ax.text(start + dur/2, y + bar_h/1.6,   # 上移一点
                    txt, ha='center', va='bottom',
                    color='black', fontsize=6, linespacing=0.9)

# ---------------- 轴、图例 ----------------
ax.set_yticks(list(y_pos.values()))
ax.set_yticklabels(threads)
ax.set_xlabel('Seconds since 21:06:19')
ax.set_title('Thread Lifecycle Gantt Chart (Auto-label placement)')

legend_handles = [Patch(facecolor=color_map[p], label=p) for p in color_map]
ax.legend(handles=legend_handles, bbox_to_anchor=(1.02, 1), loc='upper left',
          title='Phase', frameon=False)

ax.set_xlim(left=0)                     # y 方向网格可选
ax.grid(axis='x', color='#dddddd', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
